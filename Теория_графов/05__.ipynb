## I. Структурные характеристики

### 1. Построить три графа (размерностью не 100 вершин каждый): случайный граф, самоподобный граф, граф малого мира.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Создаем случайный граф с 100 вершинами и вероятностью ребра 0.1
G1 = nx.erdos_renyi_graph(100, 0.1)

# Создаем самоподобный граф с 100 вершинами и коэффициентом ассортативности 0.3
G2 = nx.barabasi_albert_graph(100, 3)

# Создаем граф малого мира с 100 вершинами, каждая вершина связана с 10 ближайшими соседями, вероятность перепрыгнуть через связь равна 0.2
G3 = nx.watts_strogatz_graph(100, 10, 0.2)

print("Случайный граф: ", nx.info(G1))
print("Самоподобный граф: ", nx.info(G2))
print("Граф малого мира: ", nx.info(G3))

plt.figure(figsize=(10, 5))

plt.subplot(131)
nx.draw(G1, node_size=10)

plt.subplot(132)
nx.draw(G2, node_size=10)

plt.subplot(133)
nx.draw(G3, node_size=10)

plt.show()
```

### 2. Для каждого построенного графа рассчитать его метрические характеристики (радиус, диаметр и т.п.).

```python
# Рассчитываем метрические характеристики для каждого графа
print("Случайный граф:")
print("Радиус: ", nx.radius(G1))
print("Диаметр: ", nx.diameter(G1))
print("Коэффициент кластеризации: ", nx.average_clustering(G1))
print("Центральность посредничества: ", nx.betweenness_centrality(G1))

print("Самоподобный граф:")
print("Радиус: ", nx.radius(G2))
print("Диаметр: ", nx.diameter(G2))
print("Коэффициент кластеризации: ", nx.average_clustering(G2))
print("Центральность посредничества: ", nx.betweenness_centrality(G2))

print("Граф малого мира:")
print("Радиус: ", nx.radius(G3))
print("Диаметр: ", nx.diameter(G3))
print("Коэффициент кластеризации: ", nx.average_clustering(G3))
print("Центральность посредничества: ", nx.betweenness_centrality(G3))
```

### 4. Построить предфрактальный граф размерностью не менее 500 вершин.

```python
def add_branches(G, depth, max_depth):
    if depth >= max_depth:
        return
    new_nodes = []
    for node in list(G.nodes()):
        for i in range(2):
            new_node = node + str(i)
            G.add_node(new_node)
            G.add_edge(node, new_node)
            new_nodes.append(new_node)
    add_branches(G, depth+1, max_depth)
    
G = nx.Graph()
G.add_node('0')
add_branches(G, 0, 9)
print(G.number_of_nodes())
```
